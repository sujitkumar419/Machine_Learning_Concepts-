```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
sns.heatmap(X.iloc[5].values.reshape(1, 13), annot=True, cmap="coolwarm")
```

```python
sns.heatmap(X.iloc[5].values.reshape(13, 1), annot=True, cmap="coolwarm")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(X,y)
```

```python

rf.feature_importances_
```

```python

rf.feature_importances_.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(rf.feature_importances_.reshape(1, -1), annot=True, cmap="coolwarm")
plt.xlabel("Features")
plt.ylabel("Importance")
plt.show()
```
