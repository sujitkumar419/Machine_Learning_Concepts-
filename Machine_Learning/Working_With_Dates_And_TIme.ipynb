```python
import pandas as pd

# Agar tumhare DataFrame ka naam time hai
time['date'] = pd.to_datetime(time['date'], errors='coerce')  # convert safely

time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

print(time.head())
```


## Extract Time part

```python
time['time'] = time['date'].dt.time

time.head()
```

```python
import pandas as pd
from datetime import datetime

# current date-time
today = datetime.today()

# ab difference nikal sakte ho
time['date'] = pd.to_datetime(time['date'])   # ensure datetime format
time['diff'] = today - time['date']

print(time.head())
```
